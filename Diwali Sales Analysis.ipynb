**Exploratory Data Analysis**

```python
df.columns
```

```python
xy = sns.countplot(x='Gender', data=df ,palette='coolwarm')

for bars in xy.containers:
    xy.bar_label(bars)
```

```python
sales_gender=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gender,palette='coolwarm')
```

*From the analysis, it is evident that females place more orders and contribute more revenue compared to males during Diwali sales*

**Top Categories**

```python
top_categories = df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False)
top_categories.plot(kind='bar', title='Revenue by Product Category')
plt.ylabel("Total Sales Amount")
plt.show()
```

*From the above graph, we can conclude that the product categories contributing the most to the total amount are Food, Clothing & Parcel, Electronics, and Footwear & Shoes. These categories show strong performance, likely due to high consumer demand and frequent purchasing patterns. In contrast, categories like Office Supplies and Veterinary Products contribute the least, indicating lower sales volumes or niche market interest.*

**Age group**

```python
df_clean = df.dropna(subset=['Age Group', 'Amount'])
df_clean['Amount'] = pd.to_numeric(df_clean['Amount'], errors='coerce')
sales_age_group = df_clean.groupby('Age Group', as_index=False)['Amount'].sum()
sales_age_group = sales_age_group.sort_values(by='Amount', ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(sales_age_group['Age Group'], sales_age_group['Amount'], color='skyblue')
plt.title('Total Order Amount by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

*from all of the age group- 26–30 age group accounts for the highest sales revenue, followed by the 36–46 age group.*

**States by total sales amount**

```python
top_state=df.groupby(['State'],as_index=False).sum().sort_values(by='Amount',ascending=False).head(10)
top_state.plot(kind='bar',x='State',y='Amount',figsize=(10,5),title='TOP 10 STATES BY TOTAL SALES AMOUNT')
plt.ylabel('TotalAmount')
plt.show()
```

*From the  graph we can say that from the top 10 states ,states like Uttar Pradesh, Maharashtra, and Karnataka generated the highest revenue during Diwali.

**Top Occupation**

```python
top_occupation=df.groupby(['Occupation'],as_index=False).sum().sort_values(by='Amount',ascending=False)
top_occupation.plot(kind='bar',x='Occupation',y='Amount',title='Spending by occupation')
plt.ylabel('Total Amount')
plt.show()
```

 From the above graph we can see top 10 sector contributing to the sales where People working in tech, healthcare, and aviation sectors spend the most during Diwali.

**Relationship between number of orders and total amount**

```python
sns.histplot(data=df,x='Orders',y='Amount',bins=30,cbar=True)
plt.title ('Orders vs Amount')
plt.show()
```

*The majority of the orders fall within the ₹5,000 to ₹10,000 range, with a sharp decline observed from ₹20,000 to ₹25,000*

**Top customers**

```python
top_customers=df.groupby('Cust_name')['Amount'].sum().sort_values(ascending=False).head(10)
top_customers.plot(kind='bar',title='top 10 customers by sales amount')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.show
```

*From the above graph we can observe that these are top10 customers,where Customers like Vishakha (₹40,000), Sudevi (~₹30,000), and Vasudev (~₹30,000) are the top contributors to sales.*

**Correlation heatmap between Age,Amount,Order**

```python


corr = df[['Amount', 'Orders', 'Age']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

*Correlation analysis reveals no strong relationship between variables such as amount, orders, and age.*
